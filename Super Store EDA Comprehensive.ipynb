# Customer Loyalty


```python
df.head(3)
```

```python
# Group data according to: Customer Name, Customer ID, Segment, and calculate the frequency of their orders
customer_order_freq = df.groupby(['Customer ID', 'Customer Name', 'Segment'])['Order ID'].count().reset_index()
customer_order_freq
```

```python
#rename order ID column to represent number of times each customer made a purchasing
customer_order_freq= customer_order_freq.rename(columns={'Order ID': 'Total Orders'})  #,inplace = True

print(customer_order_freq)
```

```python
#to rank the customers in order of frequency

repeat_customers = customer_order_freq[customer_order_freq['Total Orders'] >=1]

#sort the repeat customers in descending order
sorted_repeat_customers = repeat_customers.sort_values(by='Total Orders', ascending = False)

print(sorted_repeat_customers)
```

```python
#to print first ten only and drop the reset index
top_ten_customers = sorted_repeat_customers.head(10).reset_index(drop= True)
print(top_ten_customers)
```

# Ranking Customers According to Sales

```python
df.head(3)
```

```python
# Group data according to: Customer Name, Customer ID, Segment, and calculate total sales of each customer
customer_total_sales= df.groupby(['Customer ID', 'Customer Name', 'Segment'])['Sales'].sum().reset_index()
customer_total_sales
```

```python
#rename order ID column to represent total sales for each customer
customer_total_sales= customer_total_sales.rename(columns={'Sales': 'Total Sales'})  #,inplace = True

print(customer_total_sales)
```

# Sales Trend Analysis
### Yearly Sales Trend

```python
df.head(3)
```

```python
#convert order date to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst = True)

#grouping by year and summing the sales per year
yearly_sales = df.groupby(df['Order Date'].dt.year)['Sales'].sum().reset_index()

#rename the columns
yearly_sales = yearly_sales.rename(columns={'Order Date': 'Year', 'Sales': 'Total Sales'})
yearly_sales
```

```python
# to visualize using horizontal bar graph
plt.barh(yearly_sales['Year'], yearly_sales['Total Sales'])

#labels
plt.title('Total Yearly Sales')
plt.xlabel('Total Sales')
plt.ylabel('Year')

plt.show()
```

```python
#plotting regular bar
plt.bar(yearly_sales['Year'], yearly_sales['Total Sales'])

#labels
plt.title('Total Yearly Sales')
plt.xlabel('Year')
plt.ylabel('Total Sales')

plt.xticks(rotation =65) #to accomodate the xvalues
plt.show()
```

```python
#plotting the line graph; great for visualizing trend
plt.plot(yearly_sales['Year'], yearly_sales['Total Sales'],marker = 'o', linestyle = '--')

#labels
plt.title('Total Yearly Sales')
plt.xlabel('Year')
plt.ylabel('Total Sales')

plt.xticks(rotation =65) #to accomodate the xvalues
plt.show()
```

# Quartely Sales Trend
### 2018

```python
# convert order date to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst = True)

#filter by year
year_sales = df[df['Order Date'].dt.year == 2018]

#calculate quaterly sales for the year 2018
quarterly_sales = year_sales.resample('QE', on='Order Date')['Sales'].sum().reset_index()

#rename necessary columns
quarterly_sales = quarterly_sales.rename(columns={'Order Date': 'Quarter', 'Sales': 'Total Sales'})
print('These are the Quarterly Sales for 2018')
print(quarterly_sales)
```

```python
#plotting the line graph
plt.plot(quarterly_sales['Quarter'], quarterly_sales['Total Sales'],marker = 'o', linestyle = '--')

#labels
plt.title('Total Quartely Sales for 2018')
plt.xlabel('Quarter')
plt.ylabel('Total Sales')

plt.xticks(rotation =65) #to accomodate the xvalues
plt.show()
```

# Monthly Sales Trend for a Year
### 2018

```python
# convert Order Date to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst = True)

#filter by year
year_sales = df[df['Order Date'].dt.year == 2018]

#group by month
monthly_sales = year_sales.resample('ME', on='Order Date')['Sales'].sum().reset_index()

#rename necessary columns
monthly_sales = monthly_sales.rename(columns={'Order Date': 'Month', 'Sales': 'Total Sales'})
print('These are the Monthly sales for 2018')
print(monthly_sales)
```

```python
#plotting the line graph
plt.plot(monthly_sales['Month'], monthly_sales['Total Sales'],marker = 'o', linestyle = '--')

#labels
plt.title('Total Monthly Sales for 2018')
plt.xlabel('Month')
plt.ylabel('Total Sales')

plt.xticks(rotation =65) #to accomodate the xvalues
plt.show()
```
