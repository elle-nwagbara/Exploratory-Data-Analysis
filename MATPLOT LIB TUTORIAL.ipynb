# Plotting Graphs

```python
from matplotlib import pylab
print(pylab.__version__)
```

```python
"""Use NumPy to generate random data"""
import numpy as np
x = np.linspace(0,10,25)
y = x * x + 2
print(x)
print(y)
print(np.array([x,y]).reshape(25,2)) #gives a shape of 25 values for 2 columns each
```

```python
""" It takes only one command to draw"""
pylab.plot(x,y, 'r') # 'r' stands for red
```

# Drawing a Subgraph

```python
""" Drawing a subgraph
"""
pylab.subplot(1,2,1) #represent (rows, columns, indexes) Index is how you stack them
pylab.plot(x,y, 'r--') # Third parameter stands for colour and linestyle
pylab.subplot(1,2,2)
pylab.plot(y,x, 'g*-')
```

```python
pylab.subplot(2,1,1) #represent (rows, columns, indexes) Index is how you stack them
pylab.plot(x,y, 'r--') # Third parameter stands for colour and linestyle
pylab.subplot(2,1,2)
pylab.plot(y,x, 'g*-')
```

```python
from matplotlib import pyplot as plt
```

```python
fig = plt.figure() #the canvas

axis = fig.add_axes([0.5,0.1,0.8,0.8]) #control the left,bottom, width, and height of the convas (from 0 to 1)

axis.plot(x,y, 'r')
```

```python
# to draw subgraphs
fig, axes = plt.subplots(nrows=1, ncols=2) #submap is of 1 row and 2 columns
for ax in axes:
    ax.plot(x,y, 'r')
```

```python
# we can also draw a picture or graph inside another graph
fig = plt.figure() #the new canvas

#Control left, bottom, width, and height
axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) #big axes
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) #small axes
axes1.plot(x,y, 'r')
axes2.plot(y,x, 'b')
```

```python
#new graphic object
fig = plt.figure(figsize =(15,9), dpi=300) #fig size, dot per inches
fig.add_subplot()
plt.plot(x,y, 'r')
```

# Line colour, transparency and more in Matplotlib

```python
fig,axes = plt.subplots(dpi=152)
axes.plot(x, x*2, color = 'red', alpha=.2) #alpha is transparency
axes.plot(x, x+2, color = '#1155dd')
axes.plot(x, x+3, color = '#15cc55')
```

```python
fig,ax = plt.subplots(dpi=102)
#line width

ax.plot(x, x**2, color='red', linewidth=.25)
ax.plot(x, x*5, color='blue', linewidth =1)
ax.plot(x, x**3, color ='green', linewidth=.5)
ax.plot(x,x+1, color = 'black', linewidth=2)
```

```python
fig,ax = plt.subplots(dpi=150)
#line width

ax.plot(x, x**2, color='red', lw=2.25, linestyle='dotted')
ax.plot(x, x*5, color='blue', linewidth =1, linestyle='--')
ax.plot(x, x**3, color ='green', linewidth=1.5, ls = '-.')

line, = ax.plot(x,x+1, color = 'black', linewidth=2)
line.set_dashes([5,10,10,5])

 #'-:' is not a valid value for ls; supported values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
```

# Other 2D Graphics

```python
n = np.array([1,2,3,4,5])

fig, axes = plt.subplots(1, 4, figsize=(15,4))
axes[0].set_title("ScatterA")
axes[0].scatter(x,x+0.25 * np.random.randn(len(x)))

axes[1].set_title("Step")
axes[1].step(n,n*2, lw=2)

axes[2].set_title("Bar")
axes[2].bar(n,n**2) #align = 'center', width=0.5, alpha =0.5)

axes[3].set_title("Fill_Between")
axes[3].fill_between(x,x**2,x**3, color = 'green', alpha=0.5)
```

# Draw a radar chart

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_axes([0.0,0.0,1,1], polar=True) #polar put the degrees
t = np.linspace(0, 2 * np.pi,100)
ax.plot(t,t, color = 'blue', lw = 2)
```

# Draw a histogram

```python
n = np.random.randn(100000)
fig,axes = plt.subplots(1,2, figsize =(12,4))
axes[0].set_title('Default Histogram')
axes[0].hist(n)

axes[1].set_title('Cumulative Detailed Histogram')
axes[1].hist(n, cumulative = True, bins=10)
# Cumulative histograms are great for cumulative data visualization such as rainfall
```

# Draw Contour Image (Map)

```python
import matplotlib
import numpy as np #number arrays
import matplotlib.cm as cm #color maps
import matplotlib.pyplot as plt #plots

delta = 0.025
x = np.arange(-3.0,3.0, delta)
y = np.arange(-2.0,2.0, delta)
X,Y = np.meshgrid(x,y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X -1)**2 -(Y -1)**2)
Z = (Z1 - Z2 ) * 2

print(X)
print(Y)
```

```python
fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z)
ax.clabel(CS, inline = 1,fontsize = 10)
ax.set_title('Contour Map')
```

# Draw 3D surface images

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D

fig = plt.figure(figsize = (14,6))

#specify the 3d graphics to draw projection = '3d'
ax = fig.add_subplot(1, 2, 1, projection = '3d') #colum,width,index
ax.plot_surface(X, Y, Z, rstride=4, cstride= 4, linewidth=0)
```

# Write a python pragramming to create a pie chart of the popularity of programming languages


```python
import matplotlib.pyplot as plt

#Data to plot
languages = 'Python', 'Java', 'PHP', 'JavaScript', 'C#', 'C++'
popularity = [22.2, 17.6,8.7,8, 7.7,6.7]
colors = ['#1f77b4', '#ff7f0e', '#2ca02c','#d62728', '#9467bd', '#8c562b']
```

```python
#Explode first slice
explode = (0.1, 0, 0, 0 ,0,0)
#plot chart with attribute data
plt.pie(popularity, explode = explode, labels = languages, colors = colors,
       autopct= '%1.1f%%', shadow = True, startangle = 140) 
#autopct = auto print/picture format in python
#startangle shows where the chart (exploded piece) should begin
```
