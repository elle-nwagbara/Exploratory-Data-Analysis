```python
#to groupby continent
#df.groupby('Continent').mean().sort_values(by='2022 Population', ascending = False)

#since that didn't work, I'll create a new dataframe with the numeric values and the continent column
df_continent = df.drop(['CCA3','Country', 'Capital'],axis =1)
df_continent
```

```python
#groupby continent, find mean, and sort with 2022 Population in descending order
df_continent.groupby('Continent').mean().sort_values(by='2022 Population', ascending = False)
```

```python
#to find all values for Oceania
df[df['Continent'].str.contains('Oceania')]
```

```python
# store grouped data into new dataframe called df2
df2 = df_continent.groupby('Continent').mean().sort_values(by='2022 Population', ascending = False)
df2
```

```python
#visualize df2
df2.plot()
```

```python
#to visualize df2 where the countinents are the index
df3 = df2.transpose()
df3
```

```python
df3.plot()
```

```python
#to create a dataframe with only the population data
df2.columns
```

```python
#to create dataframe with only population by continent data
df4 = df2.drop(['Rank','Area (km²)', 'Density (per km²)',
       'Growth Rate', 'World Population Percentage'], axis = 1)
df4
```

```python
#to visualize created dataframe
df4.plot()
```

```python
#to swap data axes
df5 = df4.transpose()
df5
```

```python
#to reverse the population data so it is ordered from earliest to latest
df6 = df5[::-1]
df6
```

```python
#to visualize population trend by continent
df6.plot()
```
