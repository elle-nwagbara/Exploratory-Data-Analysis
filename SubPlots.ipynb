# Multiple Plots as SubPlots

```python
# import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
# define temp, wind, humidity, and precipitation data and Time data (Hrs)
temp_data = np.random.randint(100,size=20)
wind_data = np.random.randint(100,size=20)
hum_data = np.random.randint(100,size=20)
prec_data = np.random.randint(100,size=20)
time_hrs = np.random.randint(24,size=20)
print(temp_data)
print(wind_data)
print(hum_data)
print(prec_data)
print(time_hrs)
```

```python
# draw subplots for (1,2,1) and (1,2,2)
plt.figure(figsize=(8,4))
plt.subplots_adjust(hspace=.25)
plt.subplot(1,2,1)
plt.title('Temp(F)')
plt.plot(time_hrs,temp_data,color='b',linestyle='-',linewidth=1)
plt.subplot(1,2,2)
plt.title('Wind')
plt.plot(time_hrs,wind_data, color='r', linestyle='-',linewidth=1.5)
plt.show()
```

```python
# draw subplots for (2,1,1)
plt.figure(figsize=(6,6))
plt.subplots_adjust(hspace=.25)
plt.subplot(2,1,1)
plt.title('Humidity')
plt.plot(time_hrs,hum_data,color='b',label='humidity',linestyle='-',linewidth=1)
plt.title('Humidity & Precipitation')
plt.plot(time_hrs,prec_data, color='r', label='precipitation', linestyle='-',linewidth=1.5) 
plt.legend()
plt.show()
```

```python
# draw subplots for (2,1,1) and (2,1,2)
plt.figure(figsize=(6,6))
plt.subplots_adjust(hspace=.50)
plt.subplot(2,1,1)
plt.title('Humidity(%)')
plt.plot(time_hrs,hum_data,color='b',linestyle='-',linewidth=1)
plt.subplot(2,1,2)
plt.title('Precipitation(%)')
plt.plot(time_hrs,prec_data, color='r', linestyle='-',linewidth=1.5)
plt.show()
```

```python
# draw subplots for (2,2,1), (2,2,2), (2,2,3) and (2,2,4)
plt.figure(figsize=(9,9))
plt.subplots_adjust(hspace=.25)
plt.subplot(2,2,1)
plt.title('Temp(F)')
plt.plot(time_hrs,temp_data,color='b',linestyle='-',linewidth=1)
plt.subplot(2,2,2)
plt.title('Wind')
plt.plot(time_hrs,wind_data, color='r', linestyle='-',linewidth=1.5)
plt.subplot(2,2,3)
plt.title('Humidity(%)')
plt.plot(time_hrs,hum_data,color='g',linestyle='-',linewidth=1)
plt.subplot(2,2,4)
plt.title('Precipitation(%)')
plt.plot(time_hrs,prec_data, color='c', linestyle='-',linewidth=1.5)
plt.show()
```
