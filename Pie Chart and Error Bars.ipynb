# Pie Chart

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
job_data = ['20','40','17', '18', '5','10']
labels = 'IT', 'Marketing', 'Operations', 'Finance', 'HR', 'Sales'
explode = (0.05, 0,0,0,0,0)
plt.pie(job_data, labels = labels, explode=explode)
plt.show()
```
